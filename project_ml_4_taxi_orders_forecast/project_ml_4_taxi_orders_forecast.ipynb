# Описание проекта — Прогнозирование заказов такси
## Цели и задачи проекта

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Необходимо построить модель для такого предсказания со значением метрики RMSE на тестовой выборке не более 48. Размер тестовой выборки должен составлять 10% от исходных данных.

## Описание данных

Данные находятся в файле __taxi.csv__ со следующими признаками:
+ `datetime` — дата и время
+ `num_orders` — количество заказов

Признак __num_orders__ является целевым.

## План работы

+ Загрузим данные и выполним их ресемплирование по одному часу
+ Проанализируем данные
+ Обучим разные модели с различными гиперпараметрами
+ Проверим лучшую модель на тестовой выборке

# Загрузка, обзор и подготовка данных
## Загрузка данных

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.seasonal import seasonal_decompose

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from catboost import CatBoostRegressor
from sklearn.dummy import DummyRegressor
```

# Обучение и выбор лучшей модели
## Создание признаков

Напишем функцию для создания признаков, где:
+ Представим даты в индексации как отдельные столбцы.
+ Добавим скользящее среднее, в качестве признака.
+ Добавим «отстающие значения», в качестве признаков.

После добавления всех признаков, удалим строки с пропусками.

```python
def make_features(dt, rolling_mean_size, max_lag):
    data = dt.copy()
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['month'] = data.index.month
    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    for lag in range(1, max_lag + 1):
        data[f'lag_{lag}'] = data['num_orders'].shift(lag)
    return data

df_with_new_features = make_features(df, 10, 50)
df_with_new_features = df_with_new_features.dropna()
```

## Разбивка данных на выборки

Разбивку данных будем проводить в следующем соотношении:
+ Обучающая выборка — 70%
+ Валидационная выборка — 20%
+ Тестовая выборка — 10%

```python
features = df_with_new_features.drop(['num_orders'] , axis=1)
target = df_with_new_features['num_orders']

features_train, features_valid, target_train, target_valid = train_test_split(
    features,
    target,
    shuffle=False,
    test_size=0.3,
    random_state=12345)

features_valid, features_test, target_valid, target_test = train_test_split(
    features_valid,
    target_valid,
    shuffle=False,
    test_size=0.333,
    random_state=12345)
```

## Модель — Случайный лес

```python
# Время подбора гиперпараметров ~45 секунд

best_model = None
best_rmse = 50
best_est = 0
best_depth = 0

for est in range(1, 21, 1):
    for depth in range (1, 11):
        model = RandomForestRegressor(random_state=12345, n_estimators=est, max_depth=depth)
        model.fit(features_train, target_train)
        predicted_valid = model.predict(features_valid)
        mse = mean_squared_error(target_valid, predicted_valid)
        rmse = mse ** 0.5

        if rmse < best_rmse:
            best_model = model
            best_rmse = rmse
            best_est = est
            best_depth = depth
                   
print('Модель - Случайный лес. Лучшие показатели следующие:')
print(f'RMSE модели: {round(best_rmse)}')
print(f'Количество деревьев: {best_est}')
print(f'Глубина дерева: {best_depth}')
```

## Модель — Линейная регрессия

```python
model = LinearRegression()
model.fit(features_train, target_train)
predicted_valid = model.predict(features_valid)
mse = mean_squared_error(target_valid, predicted_valid)
rmse = round(mse ** 0.5)

print('Модель - Линейная регрессия.')
print(f'RMSE модели: {rmse}')
```

## Модель — CatBoost

```python
model = CatBoostRegressor(loss_function="RMSE", iterations=40, verbose=False)
model.fit(features_train, target_train)
predicted_valid = model.predict(features_valid)
mse = mean_squared_error(target_valid, predicted_valid)
rmse = round(mse ** 0.5)

print('Модель - CatBoost.')
print(f'RMSE модели: {rmse}')
```

## Выбор лучшей модели и проверка на тестовой выборке

Модель линейной регрессии показала лучшее значение RMSE. Проверим её на тестовой выборке.

```python
model = LinearRegression()
model.fit(features_train, target_train)
predicted_valid = model.predict(features_test)
mse = mean_squared_error(target_test, predicted_valid)
rmse = round(mse ** 0.5)

print('Модель - Линейная регрессия. Предсказание на тестовой выборке.')
print(f'RMSE модели: {rmse}')
```

Значение метрики RMSE на тестовой выборке получилось менее 48, как и требовалось в исходном задании.

## Сравнение работы лучшей и константной модели

```python
dummy_model = DummyRegressor(strategy='mean')
dummy_model.fit(features_train, target_train)
dummy_predicted = dummy_model.predict(features_test)
dummy_mse = mean_squared_error(target_test, dummy_predicted)
dummy_rmse = round(dummy_mse ** 0.5)

print(f'Константная модель. Значение RMSE: {dummy_rmse}')
```

Обычная константная модель предсказывает в 2 раза хуже, чем модель линейной регрессии.
